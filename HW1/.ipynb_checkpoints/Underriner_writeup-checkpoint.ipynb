#### 1. What types of blocks have reports of “Battery”?

As the two regressions below show, as the percentage of people in a given neighborhood living under the poverty line increase, there is an increase in the number of battery crimes. 

Conversely, as the average rent in a neighborhood increase, the numbers of batteries in the neighborhood goes down, with a large number of the high battery neighborhoods clustered in the lowest priced rental housing. 

```python
#total amount of battery per zip code  
is_battery = crimes['primary_type'] == "BATTERY"
battery = crimes[is_battery]
battery_crimes = battery.groupby(("zip", "percent_pop_poverty","percent_black_pop","percent_pop_bachelors","rent","year")).size().reset_index()
battery_crimes = battery_crimes.rename(index=str, columns={0: "number_of_battery_crimes"})
```

```python
#regress percentage of the population in poverty with the number of battery crimes 
poverty_regression = sns.regplot(x="percent_pop_poverty", y="number_of_battery_crimes", data=battery_crimes)
```

```python
#regress percentage of the population in poverty with the number of battery crimes 
rent_regression = sns.regplot(x="rent", y="number_of_battery_crimes", truncate=True, data=battery_crimes)
rent_regression.set(ylim=(0, None))
```

#### 2. What types of blocks get “Homicide”?

As the two regressions below show, there is generally a positive relationship between the number of black members of a neighborhood, and the number of murders in that neighborhood.  

Conversely, as the average rent in a neighborhood increase, the numbers of homicidies in the neighborhood goes down, with a large number of the high homicidies neighborhoods clustered in the lowest priced rental housing. 

Categorizing zip codes with over 30 murders as high murder areas, one can see that these high murder areas have a 46 percent higher poverty rate than the county average.


```python
#filter data by homicide 
is_homicide = crimes['primary_type']=="HOMICIDE"
homicide = crimes[is_homicide]
#total amount of homicide per zip code 
homicidal_crimes = homicide.groupby(("zip", "percent_pop_poverty","percent_black_pop","percent_pop_bachelors","rent","year")).size().reset_index()
homicidal_crimes = homicidal_crimes.rename(index=str, columns={0: "number_of_homicides"})
```

```python
#regress percentage of the population that is black with the number of homicides
rent_regression = sns.regplot(x="percent_black_pop", y="number_of_homicides", truncate=True, data=homicidal_crimes)
rent_regression.set(ylim=(-2, None))
```

```python
#regress percentage of the population that has a bachelors degree with the number of homicides
rent_regression = sns.regplot(x = "rent", y="number_of_homicides", truncate=True, data=homicidal_crimes)
rent_regression.set(ylim=(-2, None))
```

```python
#zip codes with over 30 murders 

is_high_homicide = homicidal_crimes['number_of_homicides'] > 30
high_homicide = homicidal_crimes[is_high_homicide]
```

```python
#areas with more than 30 homicides 
high_murder_pov = high_homicide.percent_pop_poverty.mean()
#average poverty rate over whole sample 
pov = crimes.percent_pop_poverty.mean()

#percent different 
percent_diff_murder_poverty = (((high_murder_pov - pov) / pov)*100)
```

#### A. Of the types of crimes you have data for, which crime type is the most likely given the call came from 2111 S Michigan Ave? What are the probabilities for each type of request?

This exact block is not the in data set, but 023XX S Michigan Ave is extremely close by. Its zipcode is 60616, so I looked at crime in this zip code. 

Within this, the five most likley crimes are:

Theft: 27 percent
Battery: 18 percent
Criminal Damage: 9 percent 
Deceptive Practice: 8 percent
Assault: 7 percent 




```python
#look for the block in the data set, find the total crimes in thats blocks zip code, create a columns thats 
#finds the relatively frequency of each crime. 
michigans = crimes[crimes['block'].str.contains("S Michigan Ave")]['block'].reset_index()
crimes_michigan_zip = crimes[crimes['zip'] == 60616]
total_michigan = len(crimes_michigan_zip)
mich_Crimes = crimes_michigan_zip.groupby("primary_type").size().reset_index()
mich_Crimes = mich_Crimes.rename(index=str, columns={0: "total"})
mich_Crimes["percentages"] = mich_Crimes.apply((lambda x: x.total / total_michigan), axis=1)
mich_Crimes.sort_values(by='percentages', ascending=False)
```
